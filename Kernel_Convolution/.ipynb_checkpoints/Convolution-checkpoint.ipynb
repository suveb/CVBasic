# CONVOLUTION OF KERNEL AND IMAGE

---

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

---
* The function multiplies each corresponding no. of every pixel and its neighbourhood with the corresponding element of the kernel and then sum of these value is assigned to the correponding centre position of the output pixel

* If the value of the pixel is less than 0 or greater than 255 then the value 127 is assigned(middle value)

---

```python
#Function to convolute image with kernel
def convol(img,k):
	w,h=img.shape
	f=np.zeros((w,h),np.uint8)
	a=0
	for i in range(1,w-1):
		for j in range(1,h-1):
			a=img[i-1][j-1]*k[0][0]+img[i-1][j]*k[0][1]+img[i-1][j+1]*k[0][2]+img[i][j-1]*k[1][0]+img[i][j]*k[1][1]+img[i][j+1]*k[1][2]+img[i+1][j-1]*k[2][0]+img[i+1][j]*k[2][1]+img[i+1][j+1]*k[2][2]
			if a<0 or a>255:
				f[i][j]=127
			else:
				f[i][j]=a
	return f
```

* It stores the paths of images and a kernel to sharpen the image

```python
#Storing the path of all Images in a list
imgpath = [ '..\Objects\lena_color_512.tif', '..\Objects\lbc.1.04.tiff',
            '..\Objects\l.1.01.tiff', '..\Objects\l.1.03.tiff' ]

#creating kernel for sharpen the image
k=np.array( ( [ 0, -1,  0 ],
              [-1,  5, -1 ],
              [ 0, -1,  0 ] ), np.float32)
```

* This loop convolves every image in the path and output it

```python
#For loop to traverse each image of the list
for i in range(4):

    #Reading image from the list
    img = cv2.imread( imgpath[i] ,0)

    #Taking convolution of image and kernel
    con=convol(img,k)

    #plotting origanl image and sharpen image for comparison
    #plt.subplot(4,2,2*i+1)
    plt.imshow( img ,cmap='gray' )
    plt.xticks([])
    plt.yticks([])
    plt.show()
    #plt.subplot(4,2,2*i+2)
    plt.imshow( con ,cmap='gray' )
    plt.xticks([])
    plt.yticks([])
    plt.show()


#showing image which is plotted
#plt.show()
```
