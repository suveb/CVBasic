# EDGE DETECTION USING CANNY EDGE DETECTOR

---

### ALGORITHM


**1. Gaussian Kernel 5x5 is applied to image for filtering ot any noise**

**2. Intensity gradient is calculated using L1 or L2 norm**

**3. Non-maximum suppresion is applied to the output of step 2**

**4. Using the gradient thresholds the final edge set is calculated**
                
                a. any pixel less than gradient 1 is excluded
                
                b. any pixel greater than gradient 2 is included
                
                c. for any pixel between the two gradients only the pixels directly connected to the pixel in step b are 
                
                included in the final image
---

> All the paths of images are stored in list for easy access through loop

```python
import cv2
import matplotlib.pyplot as plt


#Storing the path of all Images in a list
imgpath = [ '..\Objects\lena_color_512.tif', '..\Objects\lbc.1.04.tiff',
            '..\Objects\l.1.01.tiff', '..\Objects\l.1.03.tiff' ]
```

---
The loop traverses for each image and find the edges of the object using canny edge detector

---

```python
#For loop to traverse each image of the list
for i in range(4):

    #Reading image from the list
    img = cv2.imread(imgpath[i], 0)


    #Using canny edge detector to detect the edges from the image
    edge = cv2.Canny( img, 55, 117, L2gradient = True )


    #plotting image using matplotlib
    #plt.subplot(2,2,i+1)
    plt.imshow( img, cmap = 'gray' )
    plt.xticks([])
    plt.yticks([])
    plt.show()
    plt.imshow( edge, cmap = 'gray' )
    plt.xticks([])
    plt.yticks([])
    plt.show()
```
