# SMOOTHING AND SHARPENING THE IMAGE

---

_We use kernels to smoothen or sharpen the image by taking convolution of kernel with image_

---

> We first import all the necessary libraries need for our program

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

---
We store the path of all images in a list which can easily be accessed through loop

```python
#Storing the path of all Images in a list
imgpath = [ '..\Objects\lena_color_512.tif', '..\Objects\lbc.1.04.tiff',
            '..\Objects\l.1.01.tiff', '..\Objects\l.1.03.tiff' ]
```

---
### SMOOTHING
Here we create a kernel which when convoluted with image give Smoothen image

```python
#creating kernel for Smoothen the image
k=np.array( ( [ 1, 1, 1 ],
              [ 1, 1, 1 ],
              [ 1, 1, 1 ] ), np.float32)/9
```

---
In this loop we traverse for each element of the list
    1. We read the image from the list
    2. Image is convoluted with the kernel
    3. orignal as well as Smooth image is ploted with matplotlib

```python
#For loop to traverse each image of the list
for i in range(4):

    #Reading image from the list
    img = cv2.imread( imgpath[i] )


    #Taking convolution of image and kernel
    smooth = cv2.filter2D(img, -1, k)


    #Converting image to RGB for representing in matplotlib
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    smooth = cv2.cvtColor(smooth, cv2.COLOR_BGR2RGB)


    #plotting orignal image and sharpen image for comparison
    #plt.subplot(4,2,2*i+1)
    plt.imshow( img )
    plt.title('Orignal')
    plt.xticks([])
    plt.yticks([])
    plt.show()
    #plt.subplot(4,2,2*i+1)
    plt.imshow( smooth )
    plt.title('Smooth')
    plt.xticks([])
    plt.yticks([])
    plt.show()


#showing image which is plotted
#plt.show()
```

---
### SHARPENING
Here we create a kernel which when convoluted with image give Sharpen image

```python
#creating kernel for sharpen the image
k=np.array( ( [ 0, -1,  0 ],
              [-1,  5, -1 ],
              [ 0, -1,  0 ] ), np.float32)
```

---
In this loop we traverse for each element of the list
    1. We read the image from the list
    2. Image is convoluted with the kernel
    3. Orignal as well as Sharp image is ploted with matplotlib

```python
#For loop to traverse each image of the list
for i in range(4):

    #Reading image from the list
    img = cv2.imread(imgpath[i])


    #Taking convolution of image and kernel
    sharp = cv2.filter2D(img, -1, k)


    #Converting image to RGB for representing in matplotlib
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    sharp = cv2.cvtColor(sharp, cv2.COLOR_BGR2RGB)


    #plotting origanl image and sharpen image for comparison
    #plt.subplot(4,2,2*i+1)
    plt.imshow( img )
    plt.title('orignal')
    plt.xticks([])
    plt.yticks([])
    plt.show()
    #plt.subplot(4,2,2*i+2)
    plt.imshow( sharp )
    plt.title('Sharp')
    plt.xticks([])
    plt.yticks([])
    plt.show()
    

#showing image which is plotted
#plt.show()
```
